```python
# Importing the relevant libraries for creating the required graph
import pandas as pd
import seaborn as sns
```

```python
# Extracting the ground truth values from both TikTok and Billboard agreement files, followed by adding a new variable
# to show whether it is from TikTok or Billboard. Finally, we add these values into a dataframe named comparing_df to
# create out required graphs

tiktok_df = pd.read_excel("tt_output/agreement.xlsx")
billboard_df = pd.read_excel("bb_output/agreement.xlsx")

tt_ground_truth = {'ground truth' : tiktok_df['ground truth'], 'from' : 'tiktok'}
bb_ground_truth = {'ground truth' : billboard_df['ground truth'], 'from' : 'billboard'}

tt_ground_truth_df = pd.DataFrame(data=tt_ground_truth)
bb_ground_truth_df = pd.DataFrame(data=bb_ground_truth)

comparing_df = pd.concat([tt_ground_truth_df, bb_ground_truth_df])
```

```python
# First version of the graph (currently not used in the report)
sns.histplot(comparing_df, x = 'ground truth', hue = 'from')
```

```python
# Created a newer dataframe with a count variable such that we can create a nicer looking graph
new_comparing_df = comparing_df.value_counts().reset_index(name='count')
sns.barplot(new_comparing_df, x = 'ground truth', y = 'count', hue = 'from')
```
